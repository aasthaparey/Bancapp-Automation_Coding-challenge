## Standardizing dtypes:
standarizing the datatypes utr, status, and amountcolumns in both dataframes.




.


```python
api['utr'] = api['utr'].astype(str)
bank['utr'] = bank['utr'].astype(str)

api['status'] = api['status'].astype(str)
bank['status'] = bank['status'].astype(str)

api['amount'] = pd.to_numeric(api['amount'], errors='coerce')
bank['amount'] = pd.to_numeric(bank['amount'], errors='coerce')

api.info()
bank.info()
```
